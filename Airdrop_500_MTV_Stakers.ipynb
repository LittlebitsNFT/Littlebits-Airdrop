### Imports

```python
import requests
import datetime
import math
import random
import pickle
```

### Snapshot

```python
# date/time
print("UTC Date and Time:", datetime.datetime.now(datetime.timezone.utc).strftime("%Y-%m-%d %H:%M"), "\n")

# data fetch
url="https://e.mtv.ac/stake/tops"
payload = {'pageSize': 8500, 'pageNum': 1}
response = requests.post(url, data=payload)

# data print
print("MTV Top Stakers:")
for i, data in enumerate(response.json()['data']):
    staked_mtv = float(data['totals']) / 10**18
    staked_truncated = math.floor(staked_mtv * 10**2) / 10**2  # prevents rounding
    staked_str = '{:,.2f}'.format(staked_truncated)
    print(i+1, staked_str, data['address'])
```

### Ticket distribution

```python
staked_req_per_ticket = 5000
max_tickets = 20

tickets = []

print("Rank", "\t", "Total staked", "\t", "# tickets", "\t", "Address")
for i, data in enumerate(response.json()['data']):
    staked_mtv = float(data['totals']) / 10**18
    staked_truncated = math.floor(staked_mtv * 10**2) / 10**2  # prevents rounding
    number_of_tickets = int(staked_mtv/staked_req_per_ticket)
    number_of_tickets = number_of_tickets if number_of_tickets < max_tickets else max_tickets
    address = data['address']
    staked_str = '{:,.2f}'.format(staked_truncated)
    print(i+1, "\t", staked_str, "\t",number_of_tickets, "\t\t", address)
    for _ in range(number_of_tickets):
        tickets.append(address)
```

### Export file

```python
with open("airdrop_500_mtv_stakers_tickets_list", 'wb') as f:
    pickle.dump(tickets, f)
```

### Import file

```python
with open("airdrop_500_mtv_stakers_tickets_list", 'rb') as f:
    tickets = pickle.load(f)
```
